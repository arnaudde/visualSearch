```python
from sklearn.manifold import TSNE
import json
import os
import pickle as pk
import matplotlib.pyplot as plt
```

```python
def load_from_json(file:str):
    return json.loads(open(f'embeddings/{file}').read())
```

## Vgg 16

```python
embeddings = load_from_json('embeddings_vgg16_3000.json')
```

```python
# Function to load a batch into memory
def load_batch(data_dir, batch_id):
    with open(os.path.join(data_dir, 'data_batch_%i' % batch_id), mode='rb') as file:
        batch = pk.load(file, encoding='latin1')
    feats = batch['data'].reshape((len(batch['data']), 3, 32, 32)).transpose(0, 2, 3, 1)
    lbls = batch['labels']
    return feats, lbls

# Load the first batch
feats, labels = load_batch('cifar10', 1)
label_names = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
```

## Animal vs Machines

```python
colors = []
i = 0
for label in labels[:3000]:
    if label == 0:
        colors.append('salmon') #airplane
    elif label == 1:
        colors.append('tomato') #automobile
    elif label == 2:
        colors.append('navy') #bird
    elif label == 3:
        colors.append('blue') #cat
    elif label == 4:
        colors.append('skyblue') #deer
    elif label == 5:
        colors.append('royalblue') #dog
    elif label == 6:
        colors.append('dodgerblue') #frog
    elif label == 7:
        colors.append('steelblue') #horse
    elif label == 8:
        colors.append('red') #ship
    elif label == 9:
        colors.append('firebrick') #truck
    else:
        print(label)
    i+=1
```

```python
plt.figure(figsize = (20,20))
i = 1
for perplexity in [5,30,50,100]:
    embeddings_tSNE = TSNE(n_components=2, perplexity=perplexity).fit_transform(embeddings)
    plot = plt.subplot(int('22'+str(i)))
    plot.set_title(perplexity)
    plt.scatter(embeddings_tSNE[:,0], embeddings_tSNE[:,1], c=colors)
    i+=1
```
